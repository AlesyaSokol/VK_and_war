```python
def find_indices(list_to_check, item_to_find):
    indices = []
    for idx, value in enumerate(list_to_check):
        if value == item_to_find:
            indices.append(idx)
    return indices
```

```python
def define_topic_params(topic_ind, topics = topics):
  words = ' '.join([item[0] for item in model.get_topic(topic_ind)][:5])
  inds = find_indices(topics, topic_ind)
  count_nowar = 0
  count_z = 0
  for i in inds:
    if i<len(posts_nowar_clean):
      count_nowar += 1
    else:
      count_z += 1
  portion_nowar = count_nowar/len(posts_nowar_clean)
  # print(portion_nowar)
  portion_z = count_z/len(posts_z_clean)
  # print(portion_z)
  total_popularity = (portion_nowar + portion_z)/2
  nowar_prevalence = portion_nowar/(portion_nowar+portion_z)
  return words, total_popularity, nowar_prevalence, portion_nowar, portion_z
```

```python
def find_indices(list_to_check, item_to_find):
    indices = []
    for idx, value in enumerate(list_to_check):
        if value == item_to_find:
            indices.append(idx)
    return indices
```

```python
def define_topic_params(topic_ind, topics = topics):
  words = ' '.join([item[0] for item in model.get_topic(topic_ind)])
  inds = find_indices(topics, topic_ind)
  count = 0
  for i in inds:
      count += 1
  total_popularity = count/len(posts_clean)
  return words, total_popularity, count
```

```python
def get_random_post_topic(topic_ind, topics = topics):
  inds = find_indices(topics, topic_ind)
  random.shuffle(inds)
  texts = []
  # text = str(posts_nowar[inds[0]]['owner_id']) + ' ' + posts_nowar[inds[0]]['initial_text']
  text = posts_nowar[inds[0]]['initial_text']
  return text
```
