```python
f = open("./drive/MyDrive/Hackathon/stop_words_russian.txt", "r", encoding="utf-8")
stop_words = f.read()
f.close()
stop_words = stop_words.split("\n")
```

```python
posts_topic = []

for i in range(100):
  inds = find_indices(new_topics, i)
  posts_topic.append(' '.join([(posts_nowar_clean + posts_z_clean)[j] for j in inds]))

additional_stop_words = ['украина', 'россия', 'российский', 'украинский']

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(stop_words=stop_words, ngram_range = (1,2))
tfidf_matrix = tfidf_vectorizer.fit_transform(posts_topic)
feature_names = tfidf_vectorizer.get_feature_names_out()

import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

```python
def correct_phrase(phrase):
  words = phrase.split()
  nouns = [word for word in words if 'NOUN' in morph.parse(word)[0].tag]
  adjs = [word for word in words if 'ADJF' in morph.parse(word)[0].tag]
  nums = [word for word in words if 'NUMB' in morph.parse(word)[0].tag]
  if (len(nouns) == 1) and (len(adjs) == 1):
    try:
      parsed_noun = morph.parse(nouns[0])[0]
      gender = parsed_noun.tag.gender
      number = parsed_noun.tag.number
      case = 'nomn'
      parsed_adj = morph.parse(adjs[0])[0]
      corrected_adj = parsed_adj.inflect({case, number, gender}).word
      corrected_words = corrected_adj + ' ' + parsed_noun.word
      return corrected_words
    except:
      return phrase
  elif (len(nums) == 1) and (len(nouns) == 1):
    try:
      parsed_noun = morph.parse(nouns[0])[0]
      corrected_noun = parsed_noun.inflect({'gent', 'sing'}).word
      corrected_words = nums[0] + ' ' + corrected_noun
      return corrected_words
    except:
      return phrase
  elif (len(nouns) == 2):
    parsed_noun = morph.parse(nouns[1])[0]
    corrected_noun = parsed_noun.inflect({'gent', 'sing'}).word
    corrected_words = nouns[0] + ' ' + corrected_noun
    return corrected_words
  else:
    return phrase

def get_topic_keywords(topic_num, n_cut):
  tfidf_scores = zip(feature_names, tfidf_matrix.toarray()[topic_num])
  tfidf_scores = sorted(tfidf_scores, key=lambda x: x[1], reverse=True)
  top_tfidf_scores = tfidf_scores[:n_cut*2]
  words_initial = [top_tfidf_scores[i][0] for i in range(len(top_tfidf_scores))]
  scores = [top_tfidf_scores[i][1] for i in range(len(top_tfidf_scores))]
  # print(scores)
  words = words_initial[:2]
  for i in range(2,n_cut*2):
    if scores[i]>0.1:
      words.append(words_initial[i])

  words_remove = []

  for i in range(1,len(words)):
    if words[i] in additional_stop_words:
      words_remove.append(words[i])

  for i in range(len(words)):
    if ' ' in words[i]:
      w = words[i].split()
      for j in range(1,len(words)):
        if (words[j] in w):
          if not(words[j] in words_remove):
            words_remove.append(words[j])

  for w in words_remove:
    words.remove(w)


  words = words[:n_cut]

  for i in range(len(words)):
    if ' ' in words[i]:
      words[i] = correct_phrase(words[i])

  return words

def define_topic_params(topic_ind, topics = topics, n_cut = 6):
  # words = [item[0] for item in model.get_topic(topic_ind)]
  # words = [x for x in words if x not in stop_words]
  # if n_cut != -1:
  #   words = ' '.join(words[:n_cut])
  # else:
  #   words = ' '.join(words)
  words = get_topic_keywords(topic_ind, n_cut = n_cut)
  words = ', '.join(words)
  inds = find_indices(topics, topic_ind)
  count_nowar = 0
  count_z = 0
  for i in inds:
    if i<len(posts_nowar_clean):
      count_nowar += 1
    else:
      count_z += 1
  portion_nowar = count_nowar/len(posts_nowar_clean)
  # print(portion_nowar)
  portion_z = count_z/len(posts_z_clean)
  # print(portion_z)
  total_popularity = (portion_nowar + portion_z)/2
  nowar_prevalence = portion_nowar/(portion_nowar+portion_z)
  return words, total_popularity, nowar_prevalence, portion_nowar, portion_z
```

```python
nowar_prevs = []
total_pops = []
portions_nowar = []
portions_z = []
names = []
for i in range(100):
  words, total_popularity, nowar_prevalence, portion_nowar, portion_z = define_topic_params(i, new_topics, n_cut = 5)
  nowar_prevs.append(nowar_prevalence)
  total_pops.append(total_popularity)
  portions_nowar.append(portion_nowar)
  portions_z.append(portion_z)
  names.append(words)
  print(i, words)
```

```python
def find_indices(list_to_check, item_to_find):
    indices = []
    for idx, value in enumerate(list_to_check):
        if value == item_to_find:
            indices.append(idx)
    return indices
```

```python
posts_topic = []

for i in range(30):
  inds = find_indices(new_topics, i)
  posts_topic.append(' '.join([posts_clean[j] for j in inds]))

additional_stop_words = ['украина', 'россия', 'российский', 'украинский']

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(stop_words=stop_words, ngram_range = (1,2))
tfidf_matrix = tfidf_vectorizer.fit_transform(posts_topic)
feature_names = tfidf_vectorizer.get_feature_names_out()

import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

```python
def correct_phrase(phrase):
  words = phrase.split()
  nouns = [word for word in words if 'NOUN' in morph.parse(word)[0].tag]
  adjs = [word for word in words if 'ADJF' in morph.parse(word)[0].tag]
  nums = [word for word in words if 'NUMB' in morph.parse(word)[0].tag]
  if (len(nouns) == 1) and (len(adjs) == 1):
    try:
      parsed_noun = morph.parse(nouns[0])[0]
      gender = parsed_noun.tag.gender
      number = parsed_noun.tag.number
      case = 'nomn'
      parsed_adj = morph.parse(adjs[0])[0]
      corrected_adj = parsed_adj.inflect({case, number, gender}).word
      corrected_words = corrected_adj + ' ' + parsed_noun.word
      return corrected_words
    except:
      return phrase
  elif (len(nums) == 1) and (len(nouns) == 1):
    try:
      parsed_noun = morph.parse(nouns[0])[0]
      corrected_noun = parsed_noun.inflect({'gent', 'sing'}).word
      corrected_words = nums[0] + ' ' + corrected_noun
      return corrected_words
    except:
      return phrase
  elif (len(nouns) == 2):
    parsed_noun = morph.parse(nouns[1])[0]
    corrected_noun = parsed_noun.inflect({'gent', 'sing'}).word
    corrected_words = nouns[0] + ' ' + corrected_noun
    return corrected_words
  else:
    return phrase

def get_topic_keywords(topic_num, n_cut):
  tfidf_scores = zip(feature_names, tfidf_matrix.toarray()[topic_num])
  tfidf_scores = sorted(tfidf_scores, key=lambda x: x[1], reverse=True)
  top_tfidf_scores = tfidf_scores[:n_cut*2]
  words_initial = [top_tfidf_scores[i][0] for i in range(len(top_tfidf_scores))]
  scores = [top_tfidf_scores[i][1] for i in range(len(top_tfidf_scores))]
  # print(scores)
  words = words_initial[:2]
  for i in range(2,n_cut*2):
    if scores[i]>0.07:
      words.append(words_initial[i])

  words_remove = []

  for i in range(1,len(words)):
    if words[i] in additional_stop_words:
      words_remove.append(words[i])

  for i in range(len(words)):
    if ' ' in words[i]:
      w = words[i].split()
      for j in range(1,len(words)):
        if (words[j] in w):
          if not(words[j] in words_remove):
            words_remove.append(words[j])

  for w in words_remove:
    words.remove(w)


  words = words[:n_cut]

  for i in range(len(words)):
    if ' ' in words[i]:
      words[i] = correct_phrase(words[i])

  return words

def define_topic_params(topic_ind, topics = topics, n_cut = 6):
  # words = [item[0] for item in model.get_topic(topic_ind)]
  # words = [x for x in words if x not in stop_words]
  # if n_cut != -1:
  #   words = ' '.join(words[:n_cut])
  # else:
  #   words = ' '.join(words)
  words = get_topic_keywords(topic_ind, n_cut = n_cut)
  words = ', '.join(words)
  inds = find_indices(topics, topic_ind)
  count_z = 0
  for i in inds:
    count_z += 1
  portion_z = total_popularity/len(posts_z_clean)
  return words, total_popularity, count_z
```
